```r
library(tidyverse)
library(rvest)
```

```r
url <- "https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc"

movie_name <- url %>%
  read_html() %>%
  html_elements("h3.lister-item-header") %>%
  html_text2()

ratings <- url %>%
  read_html() %>%
  html_elements("div.ratings-imdb-rating") %>%
  html_text2() %>%
  as.numeric()

vote <- url %>%
  read_html() %>%
  html_elements("p.sort-num_votes-visible") %>%
  html_text2() 

df <- data.frame(movie_name,ratings,vote)
```

```r
# create new df with split vote
new_df <- df %>% 
  separate(col = vote, sep =  "\\|", 
          into=c("Vote","Gross","Rank"))
#check NA
new_df %>%
  is.na() %>%
  colSums()
#locate NA row
na_row <- which(is.na(new_df$Rank))
#move data and fill NA
for (i in na_row){
  new_df[i,"Rank"] <- new_df[i,"Gross"]
  new_df[i,"Gross"] <- NA
}
```

```r
genre <- url %>%
  read_html() %>%
  html_nodes("span.genre") %>%
  html_text2()

data.frame(new_df,genre)
```

```r
url1 <- "https://ceoworld.biz/2021/06/20/the-worlds-best-countries-for-quality-of-life-2021/"

best_country <- url1 %>%
  read_html() %>%
  html_elements("td.column-2") %>%
  html_text2()

score <- url1 %>%
  read_html() %>%
  html_elements("td.column-3") %>%
  html_text2()

BestCounrty_2021 <- data.frame(best_country,score)

BestCounrty_2021 <- BestCounrty_2021 %>%
  rowid_to_column() %>%
  rename(Rank = rowid)
```
